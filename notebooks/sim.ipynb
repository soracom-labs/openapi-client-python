```python
import os
import constructs

auth_key = os.environ["SORACOM_AUTH_KEY"]
auth_key_id = os.environ["SORACOM_AUTH_KEY_ID"]
config = constructs.auth(
    auth_key=auth_key,
    auth_key_id=auth_key_id,
    coverage=constructs.Coverage.JAPAN,
)
```

**References**
- [SORACOM API リファレンス / Sim](https://users.soracom.io/ja-jp/tools/api/reference/#/Sim)

```python
import soracom_api
from soracom_api.api import sim_api
from soracom_api.model.sim import Sim

with soracom_api.ApiClient(config) as api_client:
    api_instance = sim_api.SimApi(api_client)
    api_response: list[Sim] = api_instance.list_sims(limit=3)
    print(api_response)
```
